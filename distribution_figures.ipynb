```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# define a normal distribution mean and std dev
μ = 12000 # mean
σ = 2000 # std dev

# create a scipy object

sales = stats.norm(μ, σ)
```

```python
# define figure
plt.figure(figsize = (14,10))


#create subplots
plt.subplot(221)
x = np.linspace(μ - 4*σ, μ + 4 * σ, 100)
y = sales.pdf(x)
plt.plot(x,y, color = 'k')
plt.vlines([10000], 0, [sales.pdf(10000)], linestyle=':')
plt.title(f'PDF')
plt.ylabel('P(X = x)')
plt.xlabel('x')
# annotate
plt.annotate('PDF: P(X = 10,000) = 0.000121',(10000, 0.00012098), (3500, 0.000170), arrowprops={'arrowstyle': '->'})


plt.subplot(222)
top_10_percent_cutoff = sales.ppf(.9)
x = np.arange(μ - 4*σ, μ + 4*σ, 100)
y = sales.pdf(x)
plt.plot(x, y, color='black')
plt.vlines([10000], 0, [sales.pdf(10000)], linestyle=':')
plt.title(f'CDF')
x1 = np.arange(μ - 4*σ, 10000 )
y1 = sales.pdf(x1)
plt.fill_between(x1, y1, color='darkseagreen')
plt.annotate('CDF: P(X <= 10,000) = 0.158',(9300, 0.00005), (3500, 0.000170), arrowprops={'arrowstyle': '->'})
plt.ylabel('P(X = x)')
plt.xlabel('x')



plt.subplot(223)
top_10_percent_cutoff = sales.ppf(.9)
x = np.arange(μ - 4*σ, μ + 4*σ, 100)
y = sales.pdf(x)
plt.plot(x, y, color='black')
plt.vlines([15000], 0, [sales.pdf(15000)], linestyle=':')
plt.title(f'Survival Function (SF)')
x1 = np.arange( 15000,μ + 4*σ, )
y1 = sales.pdf(x1)
plt.fill_between(x1, y1, color='darkseagreen')
plt.annotate('SF: P(X > 15,000) = 0.067',(15600, 0.00002), (15000, 0.000075), arrowprops={'arrowstyle': '->'})
plt.ylabel('P(X = x)')
plt.xlabel('x')


plt.subplot(224)
top_10_percent_cutoff = sales.ppf(.9)
x = np.arange(μ - 4*σ, μ + 4*σ, 100)
y = sales.pdf(x)
plt.plot(x, y, color='black')
plt.vlines([14563], 0, [sales.pdf(14563)], linestyle=':')
# plt.vlines([μ, sales.ppf(.9)], 0, [sales.pdf(μ), sales.pdf(sales.ppf(.9))], linestyle=':')
plt.title(f'Inverse survival function (ISF)')
x1 = np.arange(top_10_percent_cutoff, μ + 4*σ)
y1 = sales.pdf(x1)
plt.fill_between(x1, y1, color='darkseagreen')
plt.annotate(f'${top_10_percent_cutoff:,.2f}', (top_10_percent_cutoff, sales.pdf(top_10_percent_cutoff)),
                 xytext=(15000, .0001), arrowprops={'arrowstyle': '->'})
plt.annotate(f'Top 10% of Sales', (15500, .000025), xytext=(15500, .00005), arrowprops={'arrowstyle': '->'})
# plt.annotate(f'Average Sales, ${μ:,.2f}', (μ, sales.pdf(μ)), xytext=(13300, .00018), arrowprops={'arrowstyle': '->'})
plt.ylabel('P(X = x)')
plt.xlabel('x')


plt.tight_layout()

#save figure if needed
# plt.savefig('figure1.svg')
```
